```python
import pandas as pd
import random
import numpy as np
```

```python
df_history = pd.read_csv("UPSC_History_MCQ_Dataset.csv")
print(df_history.head())
```

```python
name = input("Enter your name: ")
print(f"Hello, {name}! Welcome to your personalized MCQ practicer")

options = {1:"History", 2:"Georaphy", 3:"Science", 4:"Sports", 5:"Entertainment"}
choice = int(input(("Select Subject to practice by entering the sequence no. for following options:\n1. History\n2. Geography\n3. Science\n4. Sports\n5. Entertainment")))
print(f"Let's start practicing {options[choice]} questions, {name}!")
print("Press 'Exits' to stop the test and see your score at any time.")

def inputAns(user_input, x):
        if user_input == 'a':
            return df_history["Option1"][x]
        elif user_input == 'b':
            return df_history["Option2"][x]
        elif user_input == 'c':
            return df_history["Option3"][x]
        elif user_input == 'd':
            return df_history["Option4"][x]
        else:
            return None

correct = 0
incorrect = 0

for i in range(3):
    x = np.random.choice(df_history.index)
    print(f"Q] {df_history['Question'][x]}")
    print(f"A] {df_history['Option1'][x]}")
    print(f"B] {df_history['Option2'][x]}")
    print(f"C] {df_history['Option3'][x]}")
    print(f"D] {df_history['Option4'][x]}")
    print("enter 'Quit' to Quit the quiz")

    correct_option = df_history["CorrectOption"][x]
    user_input = input().lower().strip()
    if user_input == 'quit':
        break
    user_option = inputAns(user_input, x)
    if user_option is None:
        print("Invalid Option! Please enter a, b, c, or d.")
        continue

    if user_option == correct_option:
        print("Correct Answer!")
        correct += 1
    else:
        print(f"Incorrect Answer! The correct answer is: {correct_option}")
        incorrect += 1
print(f"Quiz Over! You got {correct} correct and {incorrect} incorrect answers, {name}.")

score = int((correct*2)-(incorrect*0.666))
print(f"Your total score is: {score}")
```
